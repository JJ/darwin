Using `darwin.add_logger()`, we can register a callback which receives Darwin's log messages. Here, we'll simply output them to the console.

> NOTE: the full log content for each session is also written to a file under Darwin's home directory (`$HOME/.darwin` by default)

```python
import darwin

# output Darwin's log content to console
darwin.add_logger(lambda msg: print(msg, end=''))
```

```python
# opens (or creates) a Darwin universe database
universe = darwin.open_universe('/tmp/demo.darwin')

# select a population
population = darwin.Population('neat')
population.size = 250

# select a domain
domain = darwin.Domain('unicycle')

# create a new experiment
experiment = universe.new_experiment(domain, population)

# evolution cycle
experiment.initialize_population()
for generation in range(5):
    experiment.evaluate_population()
    experiment.create_next_generation()
```
